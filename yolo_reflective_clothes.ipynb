```python
import torch

im = "./weights/test.jpg"

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
# Model
model = torch.hub.load(
    "../yolov5", "yolov5reflectiveclothes", pretrained=True, source="local"
)  # or yolov5m, yolov5l, yolov5x, custom
model = model.to(device)

# ————————————————
# 版权声明：本文为CSDN博主「青霄」的原创文章，遵循CC 4.0 BY-SA版权协议，转载请附上原文出处链接及本声明。
# 原文链接：https://blog.csdn.net/leiwuhen92/article/details/126888275
results = model(im)
results.show()
```
